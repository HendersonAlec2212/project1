# Clean the Electric Dataset

```python
# the ~ operator acts as a NOT, its an uno reverse card.
# this equations reads like this:
# new_df = old_df.loc[not in(old_df equal to 0).anyrow(axis=columns)]
# new_df =  every value in old_df NOT equal to zero, basically drop every row in all columns with value = 0
no_zeros_electric = electric_vehicle_df.loc[~(electric_vehicle_df==0).any(axis=1)]
no_zeros_electric.count()
```

```python
#verify Base MSRP has no zeros
no_zeros_electric.sort_values(by=['Base MSRP']).head()
```

```python
# This df shows all the rows with duplicate values

duplicate_list = no_zeros_electric.loc[no_zeros_electric.duplicated(subset=['Model Year', 'Model'], keep=False)]
```

```python
# I am cutting away some column I dont think we need.

no_zeros_electric = no_zeros_electric[['Model Year', 'Make', 'Model', 'Base MSRP']]

# The value_counts() will show us how many unique entries we have - yikes 6098 listings of one car..
no_zeros_electric.value_counts(subset=['Model Year', 'Model'])
```

```python
# dropping all rows with duplicate values
drop_dupes = no_zeros_electric.drop_duplicates()
```

```python
# renaming the dataframe

clean_electric_df = drop_dupes
```

```python
# checking to see if vehicles with same make but different year were preserved.
clean_electric_df.sort_values(by=['Make'])
```

```python
clean_electric_df
```

# Clean the USA Dataset

```python
usa_cars_df.head(10)
```

```python
# removing unnecessary index, and other columns
usa_cars_df = usa_cars_df.drop(['Unnamed: 0', 'color', 'state'], axis=1)
```

```python
usa_cars_df
```

```python
usa_cars_df = usa_cars_df.rename(columns={
    'year':'Model Year',
    'brand':'Make',
    'model':'Model',
    'mileage':'Mileage',
    'price':'Sold Price',
    
})
```

```python
usa_cars_df = usa_cars_df[['Model Year','Make','Model','Sold Price','Mileage']]
usa_cars_df
```

```python
# check for duplicates
usa_cars_df.value_counts()
```

```python
# making a dataframe of only duplicate values
dupe_check_usa = usa_cars_df.loc[usa_cars_df.duplicated(subset=['Make', 'Mileage'],keep=False)]
dupe_check_usa
```

```python
clean_usa = usa_cars_df.drop_duplicates()
```

```python
usa_cars_df.count()
```

```python
clean_usa.count()
```

# Save as new csv before merge

```python
clean_usa.to_csv(clean_usa_path)
clean_electric_df.to_csv(clean_electric_path)
```

```python
combined_df = pd.merge(clean_usa, clean_electric_df, how = 'outer')
```

```python
combined_df.head(20)
```

```python
merged_usa_electric = combined_df.fillna(0)
merged_usa_electric
```
