```python
import pandas as pd
import numpy as np
import scipy.stats as st
import os
import matplotlib.pyplot as plt
os.getcwd()
```

```python
inpath = 'raw_csv_files\\sedan_'
file_number = 'total'
outpath = 'output_data\\alec\\'
```

```python
sedan_df = pd.read_csv(f'{inpath}{file_number}.csv')
column_names = sedan_df.columns.to_list()
sedan_df = sedan_df.fillna(0)
sedan_df = sedan_df.rename(columns={
    'vf_BodyClass':'Body Class',
    'brandName':'Brand',
    'vf_Make':'Make',
    'modelName':'Model',
    'vf_Model':'Model_x',
    'vf_ModelID':'Model_ID',
    'vf_ModelYear':'Year',
    'mileage':'Mileage',
    'askPrice':'Ask Price',
    'msrp':'MSRP',
    'vf_BasePrice':'Base Price',
    'askPrice.1':'Ask Price_x',
    'color':'Color',
    'interiorColor':'Interior Color',
    'vf_BodyCabType':'Body Cab Type',
    'vf_Doors':'Doors',
    'vf_EngineCylinders':'EngineCylinders',
    'vf_EngineHP':'EngineHP',
    'vf_FuelTypePrimary':'FuelTypePrimary',
    'vf_FuelTypeSecondary':'FuelTypeSecondary',
    'vf_Series':'Series',
    'vf_Series2':'Series2',
    'vf_SteeringLocation':'SteeringLocation',
    'vf_TransmissionStyle':'TransmissionStyle',
    'vf_Trim':'Trim',
    'vf_Trim2':'Trim2',
    'vf_Turbo':'Turbo',
    'vin':'VIN',})

sedan_df.describe()
```

```python
# eliminate cars w/o a price since we will be leaning on price as main comparator
clean_sedan = sedan_df.loc[(sedan_df['Ask Price']>0) & (sedan_df['Ask Price']<200000) & (sedan_df['Mileage']<300000)]

miles_per_year=(clean_sedan['Mileage'] / (2020 - clean_sedan['Year']))

car_age = (2020-clean_sedan['Year'])

clean_sedan.insert(6, 'Vehicle Age', car_age)
clean_sedan.insert(8, 'Miles Driven Per Year', miles_per_year)


clean_sedan = clean_sedan.drop(['Body Class', 'Brand','Model_x', 'MSRP', 'Ask Price_x'], axis=1)

clean_sedan.head(20)
```

```python
# reset the index drop old index to avoid added as new column
clean_sedan = clean_sedan.reset_index(drop=True)

# clean_sedan.to_csv(f'{outpath}clean_sedan.csv', index=False)


column_names = clean_sedan.columns.to_list()
```

```python
avg_cost = clean_sedan.groupby('Make').mean()['Ask Price']
avg_mileage = clean_sedan.groupby('Make').mean()['Mileage']
sum_of_make = clean_sedan['Make'].value_counts()
avg_age = clean_sedan.groupby('Make').mean()['Vehicle Age']

# most pop for least amount best bang for buck

manufacturer_summary = pd.DataFrame({
    'Total Vehicles': sum_of_make,
    'Avg Cost': avg_cost,
    'Avg Mileage':avg_mileage,
    'Avg Age in Yrs':avg_age
    
})


manufacturer_summary.sort_values(by=['Avg Cost'])

manufacturer_summary.astype({
    'Avg Cost': 'int32',
    
})

# manufacturer_summary.to_csv(f'{outpath}sedan_manufacturer_summary.csv', index=False)
```

# Bar Chart - Amnt & avg price

```python
make_list = clean_sedan['Make'].value_counts().index.to_list()
make_amt = clean_sedan['Make'].value_counts().to_list()
avg_price_make = clean_sedan.groupby('Make').mean()['Mileage'].to_dict()
avg_price_make_list = [avg_price_make[car] for car in make_list]


fig = plt.figure(figsize=(19.20,10.80))
ax = plt.subplot(111)


avg_cost_all = clean_sedan['Mileage'].mean()
Avg = round(avg_cost_all,2)
ax.hlines(avg_cost_all, -1, 20, colors='b', linestyles='dashed', label='Avg Mileage, Inclusive')
plt.text(0.65,0.31,str(Avg), horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)

x1 = np.arange(len(make_list[0:19]))
y1 = make_amt[0:19]
y2 = avg_price_make_list[0:19]

plt.xlabel('Manufacturers')

w = 0.3
tick_locations = [value+0.4 for value in x1]
bar1 = ax.bar(x1,y1, color='c', edgecolor='k', label='Amt on Market', width=w,)
bar2 = ax.bar(x1+w,y2, color='g', edgecolor='k', label='Avg Price, Exclusive',width=w)
ax.legend()
plt.xticks(tick_locations, make_list[0:19], rotation=90, )
plt.title('Manufacturers: Quantity on Market & Average Cost')
plt.rcParams['axes.axisbelow'] = True
plt.grid(True)

# plt.savefig(f'{outpath}Sedan Quantity on Market & Average Cost.png')

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
corr = clean_sedan.corr()
```

```python
# Linear Regression Using SKLearn

#1. Determine X & y

y = clean_sedan['Ask Price'].values
x = clean_sedan['Mileage'].values


#2. Split the Dataset in Training Set and Test Set

x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=.30,random_state=0)

# Visualize Data Split
fig = plt.figure(figsize=(19.20,10.80))
plt.scatter(x_train,y_train, label='Training Data', color='r',alpha=.7)
plt.scatter(x_test,y_test, label='Testing Data', color='g',alpha=.7)
plt.legend()
```

The trend of this graph appears to move outside the first dimension, possibly a function of 1/x.

ACurved polynomial would better represent the trend as the price never crosses the x or y axis

```python
#3. Train the Model on the Training Set

LR =LinearRegression().fit(x_train.reshape(-1,1),y_train)

# Use Model to Predict on Test Data

prediction = LR.predict(x_test.reshape(-1,1))

# Plot Prediction Line Against Actual Test Data
fig = plt.figure(figsize=(19.20,10.80))
# plt.plot(x_test, prediction, label = 'Linear Regression',color='g')

plt.plot()
plt.scatter(x_test,y_test,label = 'Actual Test Data',alpha=.7,edgecolors='k')
plt.title('Mileage vs Ask Price')
plt.xlabel('Mileage')
plt.ylabel('Ask Price')

avg_cost_all = clean_sedan['Ask Price'].mean()

avg_mileage_all = clean_sedan['Mileage'].mean()

x_min = clean_sedan['Ask Price'].min()
x_max = clean_sedan['Ask Price'].max()
plt.vlines(avg_mileage_all, x_min, x_max, colors='m', linestyles='dashed', label='Avg Mileage, Inclusive')
y_min = clean_sedan['Mileage'].min()
y_max = clean_sedan['Mileage'].max()
plt.hlines(avg_cost_all, y_min, y_max, colors='r', linestyles='dashed', label='Avg Price, Inclusive')

plt.legend()
plt.grid(True)

# Evaluate the Model
scr = LR.score(x_train.reshape(-1,1),y_train)
scr = '{:,.2f}%'.format(scr*100)
print('The Overall Accuracry of the Model is: '+scr)

r2 = r2_score(y_test,prediction)
print('The R^2 Value of the Linear Regression Model is: '+str(r2))


m= float(LR.coef_)
m=  '{:,.2f}'.format(m)
b= float(LR.intercept_)
b=  '{:,.2f}'.format(b)

print('The Linear Equation of the Regression is: Y='+str(m)+'X + '+str(b))
plt.savefig(f'{outpath} Price v Mileage.png')
plt.show()
```
